```python
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
display(melb_data.head())
#display(melb_data.tail)
```

```python
melb_df = melb_data.copy()
melb_df.head()
```

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.info()
```

```python
customer_df = pd.DataFrame({
            'number': [0, 1, 2, 3, 4],
            'cust_id': [128, 1201, 9832, 4392, 7472],
            'cust_age': [13, 21, 19, 21, 60],
            'cust_sale': [0, 0, 0.2, 0.15, 0.3],
            'cust_year_birth': [2008, 2000, 2002, 2000, 1961],
            'cust_order': [1400, 14142, 900, 1240, 8430]
        })


def delete_colums(df, col=[]):
    if set(col).issubset(df.columns) is False:
        return None
    return df.drop(col, axis=1)

display(delete_colums(customer_df, ['number','cust_sale']))
```

```python
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'area': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})

countries_df['density'] = countries_df['population'] * 10**6 / countries_df['area']
display(round(countries_df['density'].mean(), 2))
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
display(melb_df['Date'])
```

```python
melb_df['WeekdaySale'] = melb_df['Date'].dt.day_of_week
display(melb_df['WeekdaySale'])
weekend_count = len(melb_df.loc[(melb_df['WeekdaySale']==5)|(melb_df['WeekdaySale']==6)])
display(weekend_count)
```

```python
def get_weekend(weekday):
    return 1 if weekday == 5 or weekday == 6 else 0

melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)
display(melb_df['Price'][melb_df['Weekend'] == 1].mean())
```

```python
popular_sellers = melb_df['SellerG'].value_counts().nlargest(49).index
melb_df['Popular_seller'] = melb_df['SellerG'].apply(lambda x: x if x in popular_sellers else 'other')
display(round(melb_df['Price'][melb_df['Popular_seller'] == 'Nelson'].min()
                    /melb_df['Price'][melb_df['Popular_seller'] == 'other'].min(), ndigits=1))
```

```python
#import pandas as pd
#melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')

# Шаг 1. Узнаём текущий объём памяти
memory_before = melb_df.memory_usage(deep=True).sum() / 1024**2
print(f"Объём памяти до преобразования: {memory_before:.1f} Мб")

# Шаг 2. Преобразование столбца Suburb
top_119_suburbs = melb_df['Suburb'].value_counts().nlargest(119).index
melb_df['Suburb'] = melb_df['Suburb'].apply(lambda x: x if x in top_119_suburbs else 'other')
melb_df['Suburb'] = melb_df['Suburb'].astype('category')

# Шаг 3. Узнаём объём памяти после преобразования
memory_after = melb_df.memory_usage(deep=True).sum() / 1024 ** 2
print(f"Объём памяти после преобразования: {memory_after:.1f} Мб")

# Разница в объёме памяти
memory_difference = memory_before - memory_after
print(f"Разница в объёме памяти: {memory_difference:.1f} Мб")
```
