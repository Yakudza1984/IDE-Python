```python
import pandas as pd
movies_df = pd.read_csv('data/movies_data/movies.csv')
display(movies_df)
```

```python
movies_df.groupby('title').count()
```

```python
users1_df = pd.read_csv('data/movies_data/ratings1.csv')
users1_df.groupby('userId').count()
```

```python
dates_df = pd.read_csv('data/movies_data/dates.csv')
dates_df['date'] = pd.to_datetime(dates_df['date'])
print(dates_df['date'].dt.year.value_counts().sort_values(ascending=False))
```

```python
users2_df = pd.read_csv('data/movies_data/ratings2.csv')
ratings_df = pd.concat([users1_df, users2_df], ignore_index=True)
display(ratings_df)
```

```python
ratings = ratings_df.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates_df], axis=1)
display(ratings_dates.tail(7))
```

```python
import os

#path = 'data/movies_data/'
def concat_user_files(path):
    rez = sorted(os.listdir(path))
    users_df = pd.read_csv(os.path.join(path, rez[0]))
    for item in rez:
        item = os.path.join(path, item)
        users = pd.read_csv(item)    
        users_df = pd.concat([users_df, users])
    
    users_df = users_df.drop_duplicates(ignore_index=True)
    return users_df    
    
print(concat_user_files('data/movies_data/'))
```

```python
joined_false = ratings_dates.join(
    movies_df,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
merged = ratings_dates.merge(
    movies_df,
    on='movieId',
    how='left'
)
print(merged[merged['rating']==5])
```
