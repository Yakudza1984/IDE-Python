```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
display(melb_df['Date'])
```

```python
print(melb_df['Date'].dt.quarter.value_counts().index[1])
print(melb_df[melb_df['Date'].dt.quarter == 2].shape[0])
```

```python
columns_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for col in melb_df.columns:
    if col not in columns_to_exclude and melb_df[col].nunique()<150:
        melb_df[col] = melb_df[col].astype('category')
        
melb_df.select_dtypes(include="category")
```

```python
melb_df.sort_values(by='Price').head(10)
```

```python
melb_df.sort_values(by='Date', ascending=False)
```

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(
    by='AreaRatio',
    ascending=True,
    ignore_index=True
)

print(melb_df['BuildingArea'][1557])
```

```python
mask1 = melb_df['Type'] == 'townhouse'
mask2 = melb_df['Rooms'] > 2
filtered_data = melb_df[mask1 & mask2].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
    )
print(filtered_data.loc[18, 'Price'])
```
